```python
import numpy as np
import pandas as pd

from scipy.stats import chi2_contingency

from twitter_geo.state_word_counts.models import StateCount
```

```python
token = 'god'
a = StateCount.state_token_count('AL', token)
b = StateCount.state_token_count('CA', token)
c = StateCount.state_count('AL')
d = StateCount.state_count('CA')
```

```python
chi2_contingency([[a, b], [c, d]])
```

```python
c = StateCount.state_count('CA')
d = StateCount.state_count('MA')

data = []
for token in StateCount.tokens():
    
    a = StateCount.state_token_count('CA', token)
    b = StateCount.state_token_count('MA', token)
    
    if a and b:
        chi, p, _, _ = chi2_contingency([[a, b], [c, d]])
        data.append((token, a/c - b/d))
```

```python
df = pd.DataFrame(data, columns=('token', 'zeta'))
```
