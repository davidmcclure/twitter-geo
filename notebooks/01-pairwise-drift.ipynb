```python
from gensim.models import Word2Vec
```

```python
al = Word2Vec.load('../models/AL.bin')
```

```python
ma = Word2Vec.load('../models/MA.bin')
```

```python
al.wv.most_similar(['money'])
```

```python
ma.wv.most_similar(['money'])
```

```python
from wordfreq import top_n_list
```

```python
words = set(top_n_list('en', 1000))
```

```python
ma_vocab = set(ma.wv.vocab.keys())
```

```python
al_vocab = set(al.wv.vocab.keys())
```

```python
vocab = set.intersection(ma_vocab, al_vocab, words)
```

```python
len(vocab)
```

```python
from itertools import combinations
```

```python
pairs = list(combinations(vocab, 2))
```

```python
ma_d = {}
```

```python
for w1, w2 in pairs:
    ma_d[w1, w2] = ma.wv.similarity(w1, w2)
```

```python
al_d = {}
```

```python
for w1, w2 in pairs:
    al_d[w1, w2] = al.wv.similarity(w1, w2)
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
data = []
for w1, w2 in ma_d.keys():
    data.append((w1, w2, ma_d[w1, w2] - al_d[w1, w2]))
```

```python
df = pd.DataFrame(data, columns=('w1', 'w2', 'delta'))
```

# Close in MA, not in AL
