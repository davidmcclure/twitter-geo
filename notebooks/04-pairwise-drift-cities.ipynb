```python
from gensim.models import Word2Vec
```

```python
bos = Word2Vec.load('../models/boston.bin')
```

```python
sf = Word2Vec.load('../models/sf.bin')
```

```python
bos.wv.most_similar(['money'])
```

```python
sf.wv.most_similar(['money'])
```

```python
from wordfreq import top_n_list
```

```python
words = set(top_n_list('en', 1000))
```

```python
bos_vocab = set(bos.wv.vocab.keys())
```

```python
sf_vocab = set(sf.wv.vocab.keys())
```

```python
vocab = set.intersection(bos_vocab, sf_vocab, words)
```

```python
len(vocab)
```

```python
from itertools import combinations
```

```python
pairs = list(combinations(vocab, 2))
```

```python
bos_d = {}
```

```python
for w1, w2 in pairs:
    bos_d[w1, w2] = bos.wv.similarity(w1, w2)
```

```python
sf_d = {}
```

```python
for w1, w2 in pairs:
    sf_d[w1, w2] = sf.wv.similarity(w1, w2)
```

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
data = []
for w1, w2 in bos_d.keys():
    data.append((w1, w2, bos_d[w1, w2] - sf_d[w1, w2]))
```

```python
df = pd.DataFrame(data, columns=('w1', 'w2', 'delta'))
```
