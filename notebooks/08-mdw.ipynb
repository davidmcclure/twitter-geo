```python
import pandas as pd

from wordfreq import top_n_list

from twitter_geo.state_word_counts.db import StateWordCount
```

```python
StateWordCount.state_token_rel_count('AL', 'gun')
```

```python
StateWordCount.state_token_rel_count('MA', 'gun')
```

```python
tokens = StateWordCount.tokens()
```

```python
topn = set(top_n_list('en', 300))
```

```python
data = []

for token in StateWordCount.tokens():
    
    if token in topn:
        continue
    
    try:
        c1 = StateWordCount.state_token_rel_count('TX', token)
        c2 = StateWordCount.state_token_rel_count('AL', token)
        data.append((token, c1-c2))
    except: pass
```

```python
df = pd.DataFrame(data, columns=('token', 'score'))
```
